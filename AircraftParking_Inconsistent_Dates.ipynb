```python
import pandas as pd

def find_date_column(df):
    possible_date_cols = []

    for col in df.columns:
        try:
            parsed = pd.to_datetime(df[col], errors='coerce')  # Convert to datetime
            valid_dates = parsed.notna().sum() / len(df)  # Percentage of valid entries

            # Ensure column was originally non-numeric before conversion
            is_object_or_string = df[col].dtype == "object" or pd.api.types.is_string_dtype(df[col])

            print(f"🔎 Checking '{col}': {valid_dates:.2%} valid dates | Type: {df[col].dtype}")  # Debugging

            if valid_dates > 0.6 and is_object_or_string:  # At least 60% valid dates & originally text-based
                possible_date_cols.append(col)
        except Exception as e:
            print(f"⚠️ Error processing '{col}': {e}")

    return possible_date_cols[0] if possible_date_cols else None

# Example usage:
date_col = find_date_column(df)
print(f"📅 Detected date column: {date_col}" if date_col else "❌ No date column detected!")
```

```python
df.loc[df.index[-50:], date_col] = pd.to_datetime(df[date_col], errors='coerce').dt.strftime('%B %d, %Y')

print(df)

count_written = df[date_col].astype(str).str.match(r'^[A-Za-z]+ \d{1,2}, \d{4}$').sum()
print(f"{count_written} dates are in written format! This data is inconsistent. Consider makeing entire {date_col} column the same format.")
```

```python
def format_date_columns(df, date_cols, format_string):
    # Ensure date_cols is a list even if a single column name is passed as a string
    if isinstance(date_cols, str):
        date_cols = [date_cols]

    for col in date_cols:
      # Use errors='coerce' in to_datetime to handle potential parsing issues
      df[col] = pd.to_datetime(df[col], errors='coerce', format='mixed').dt.strftime(format_string)
    return df

format_string = '%Y-%m-%d' # Desired output format

# Pass date_col as a list containing the column name
df = format_date_columns(df, date_col, format_string)
print(df)

count_written = df[date_col].astype(str).str.match(r'^[A-Za-z]+ \d{1,2}, \d{4}$').sum()

print(f"{count_written} dates are in written format! {date_col} column is now in the same format. This data is now consistent.")
```
